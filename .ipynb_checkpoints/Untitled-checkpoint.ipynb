```python

import requests
import pandas as pd
from bs4 import BeautifulSoup
 
url = 'http://www.redditsoccerstreams.tv/'
reqs = requests.get(url)
soup = BeautifulSoup(reqs.text, 'html.parser')

Matches = []
Links = []
for link in soup.find_all('td',class_="et3"):
    htmlval = link.find_next_siblings('td')
    soupMatch = BeautifulSoup(str(htmlval[0]))
    soupLink = BeautifulSoup(str(htmlval[1]))
    if(soup.get_text() and len(soupLink.find_all('a', href=True))):
        Matches.append(soupMatch.get_text().lower())
        UrlValue = soupLink.find_all('a', href=True)[0]['href']
        reqs = requests.get(UrlValue)
        soup = BeautifulSoup(reqs.text, 'html.parser')
        if(len(soup.find_all('iframe'))):
            Links.append(soup.find_all('iframe')[0]['src'])
        
```

```python
dataFrame = pd.DataFrame([Matches,Links],).T
dataFrame.columns = ["Match","Link"]
dataFrame
```

```python
import pandas as pd
import matplotlib.pyplot as plt
class Category:
    ##write your email here for eg:email = "thivinanandh@iisc.ac.in"
    email = "gurumurti@iisc.ac.in"
    ##TODO
    
    def __init__(self, name: str):
        self.name = name
        self.ledger = []
        self.fund = 0.0
        self.df = pd.DataFrame(self.ledger, columns=["amount", "description"])

    def check_fund(self, amount: float):
        # Check if sufficient funds are available
        assert amount >= 0, f"Amount  {amount} should not be negative"
        if self.fund<amount:
            return False
        else:
            return True

    def deposit(self, amount: float, description=""):
        #Deposit amount and print description
        assert amount >= 0, f"Amount  {amount} should not be negative"
        f = {"amount": amount, "description": description}
        self.ledger.append(f)
        self.fund += amount

    def withdraw(self, amount: float, description=""):
        #Withdraw amount and print description
        assert amount >= 0, f"Amount  {amount} should not be negative"
        if self.check_fund(amount):
            #f = {"amount": {amount}, "description": {description}}
            f = {"amount": amount*-1.0, "description": description}
            self.ledger.append(f)
            self.fund -= amount
            return True
        else:
            return False

    def get_balance(self):
        return self.fund


    def transfer(self,amount: float, category: str):
        #Transfer funds from one category to another
        if self.check_fund(amount):
            self.withdraw(amount, f"Transfer to {category.name}")
            category.deposit(amount, f"Transfer from {self.name}")
            return True
        else:
            return False

    def print_class(self):
        #print(self.ledger)
        self.df = pd.DataFrame(self.ledger, columns=["amount", "description"])
        final_print = f"{self.name.center(30,'*')}\n"
        for i in range(0, self.df.shape[0]):
            desc = str(self.df['description'][i][0:23]).ljust(23)
            amt = self.df['amount'][i]
            f_amt = my_formatter.format(amt)
            final_print += f"{desc:>22}{f_amt:>7}\n"
        final_print += f"Total: {self.fund}"
        return final_print

    def __repr__(self):
        return self.print_class()

    def get_df(self):
        self.df = pd.DataFrame(self.ledger, columns=["amount", "description"])
        return self.df
    
def create_spend_chart(categories):
    ##TODO
    categories = []
    withdrawal = []
    for l in List:
        df = l.get_df()
        df1 = df[df['amount'] < 0 ]
        w_sum = -1.0*df1['amount'].sum()
        #w_sum1 = round(w_sum/10)*10
        categories.append(l.name.lower().capitalize())
        withdrawal.append(w_sum)

    percent_withdrawal = [round(i / sum(withdrawal) * 10)*10 for i in withdrawal]
    graphString = 'Percentage spent by category\n'
    for i in range(100, -10, -10):
        graphString += str(i).rjust(3) + "| "
        for percent in percent_withdrawal:
            graphString += "o  " if percent >= i else "   "
        graphString += "\n"

    graphString += ' ' * 4 + '-' * (2 * (len(categories) + 1) + 2)
    max_len = len(max(categories, key=len))
    categories = [i.ljust(max_len) for i in categories]
    for i in range(max_len):
        graphString += '\n     '
        for cat in categories:
            graphString += cat[i] + '  '


    return graphString;
```
